# Jupyter Notebook
***
Markdown is used to create non-code text contents in Jupyter Notebook. Select Markdown from dropdown menu.


Markdown syntaxs:
Heading:
Use #(space) to create heading with largest heading size.
Keep on adding ##(space) to create heading with smaller font size.
Examples:


Horizontal line separator:
*** is used as horizontal line separator.
Example: 

Text indentation:
Use >(space)(text) for text indentation.
Example: 

Monospace font:
Monospace font is some text highlighted in a light gray background. Typically used to indicate file paths, user entered messages, file names etc,.
Example: `this is monospace font`
Example:


Mathematical symbols
Mathematical symbols can be shown between two $ symbols.

Example: $x = x + y$



>Line break:
Use <"br">(without quotes) for line breaks.
Example:

>Bold:
Use double underscore(text)double underscore {__text__} to bold the text.
Example: 
[Italics]:
Use single underscore(text)single underscore {_text_} to italics the text.
Example:
Font color:
Use "<font color=red>Text<font color>"

Numbered list and sub-list:
    
1. the (number.(space) to number)
    1. sub text
    
* (for bulletins)


Keyboard shortcuts
